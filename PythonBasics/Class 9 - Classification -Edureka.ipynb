```python
import pandas as pd
df = pd.read_csv('Social_Network_Ads.csv')
```

```python
x = df.iloc[:,2:4]
```

```python
y = df.iloc[:,4].values
```

```python
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
x = sc_x.fit_transform(x)
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test, y_train, y_test = train_test_split(x,y,test_size=30)
```

```python
from sklearn.linear_model import LogisticRegression
logmodel = LogisticRegression()
logmodel.fit(x_train,y_train)
```

```python
x_test
```

```python
y_test
```

```python
y_pred_logmodel = logmodel.predict(x_test)
```

```python
y_pred_logmodel
```

```python
#KNN
from sklearn.neighbors import KNeighborsClassifier
classifier_knn = KNeighborsClassifier(n_neighbors=5,p=2)
classifier_knn.fit(x_train,y_train)
```

```python
y_pred_knn = classifier_knn.predict(x_test)
```

```python
y_pred_knn
```

```python
#Naive Bayes
from sklearn.naive_bayes import GaussianNB
classifier_nb = GaussianNB()
classifier_nb.fit(x_train,y_train)
```

```python
y_pred_nb = classifier_nb.predict(x_test)
```

```python
#SVM
from sklearn.svm import SVC
classifier_svm_sig = SVC(kernel='sigmoid')
classifier_svm_sig.fit(x_train,y_train)
```

```python
y_pred_svm_sig = classifier_svm_sig.predict(x_test)
```

```python
#SVM
from sklearn.svm import SVC
classifier_svm_lin = SVC(kernel='linear')
classifier_svm_lin.fit(x_train,y_train)
```

```python
y_pred_svm_lin = classifier_svm_lin.predict(x_test)
```

```python
y_pred_svm_lin
```

```python
#SVM
from sklearn.svm import SVC
classifier_svm_rbf = SVC(kernel='rbf')
classifier_svm_rbf.fit(x_train,y_train)
```

```python
y_pred_svm_rbf = classifier_svm_rbf.predict(x_test)
```

```python
y_pred_svm_rbf
```

```python
#DT
from sklearn.tree import DecisionTreeClassifier
classifier_dt = DecisionTreeClassifier(criterion='entropy')
classifier_dt.fit(x_train, y_train)
```

```python
y_pred_dt = classifier_dt.predict(x_test)
```

```python
#RF
from sklearn.ensemble import RandomForestClassifier
classifier_rf = RandomForestClassifier(n_estimators=3, criterion='entropy')
classifier_rf.fit(x_train,y_train)
```

```python
y_pred_rf = classifier_rf.predict(x_test)
```
